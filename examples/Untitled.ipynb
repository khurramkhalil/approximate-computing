```python
import time
import psutil

while True:
    # Get the percentage of CPU usage
    cpu_percentage = psutil.cpu_percent(interval=1)
    print(f"CPU Utilization: {cpu_percentage}%")
    time.sleep(1)
```

```python
import subprocess

def get_cpu_utilization():
    result = subprocess.run(['top', '-bn1'], stdout=subprocess.PIPE)
    output = result.stdout.decode('utf-8')
    
    # Find CPU usage in the top command output
    lines = output.splitlines()
    for line in lines:
        if 'Cpu(s)' in line:
            cpu_usage = line.split(':')[1].split(',')[0]
            return cpu_usage.strip()

while True:
    
    cpu_utilization = get_cpu_utilization()
    print(f"CPU Utilization: {cpu_utilization}")
    time.sleep(2)
```
